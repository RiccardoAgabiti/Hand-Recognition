```python
import cv2
import mediapipe as mp

# Inizializza MediaPipe Hands con parametri aggiornati per una maggiore precisione
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False,
                       max_num_hands=2, # Modifica il massimo numero di mani
                       min_detection_confidence=0.7,  # Aumenta la confidenza della rilevazione
                       min_tracking_confidence=0.7)  # Aumenta la confidenza del tracciamento

# Inizializza MediaPipe Drawing
mp_drawing = mp.solutions.drawing_utils

# Cattura video dalla webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Errore: Impossibile accedere alla webcam.")
    exit()

while cap.isOpened():
    success, image = cap.read()
    if not success:
        print("Ignorando frame vuoto dalla webcam.")
        continue

    # Ruota l'immagine allo specchio prima del processamento
    image = cv2.flip(image, 1)

    # Converti l'immagine da BGR a RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Processa l'immagine specchiata e trova le mani
    results = hands.process(image_rgb)

    # Disegna le mani sull'immagine specchiata
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)

    # Riporta l'immagine alle dimensioni iniziali (non necessario se l'immagine non è stata ridimensionata dopo il ribaltamento)
    # image = cv2.resize(image, (640, 480))  # Rimuovere o commentare se le dimensioni originali sono mantenute

    # Mostra l'immagine ribaltata e con le mani annotate
    cv2.imshow('Riconoscimento Mani con MediaPipe', image)

    # Premere 'q' per uscire
    if cv2.waitKey(5) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
